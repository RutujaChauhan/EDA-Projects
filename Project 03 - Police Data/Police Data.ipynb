Q. 2) Question ( Based on Filtering + Value Counts ) - For Speeding , were Men or Women stopped more often ? 

```python
#df[df.violation_raw == "Speeding"]["driver_gender"].value_counts()
df[df.violation_raw == "Speeding"].groupby(by="driver_gender").count()["stop_date"]
```

Q. 3) Question ( Groupby ) - Does gender affect who gets searched during a stop ?
Question ( mapping + data-type casting )

```python
df.groupby(by="driver_gender")["search_conducted"].sum()
```

Q. 4) Question ( mapping + data-type casting ) - What is the mean stop_duration ?

```python
df["sstop_duration"] = df.stop_duration.map(lambda x: 7.5 if x == "0-15 Min" else 22.5 if x == "16-30 Min" else 45 if x == "30+ Min" else 0)
```

```python
df.sstop_duration.mean()
```
